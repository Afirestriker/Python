```python
# 1: Create a basic Box Plot with Title and label to x-axis and y-axis

import numpy as np
import matplotlib.pyplot as plt

sales = np.array([62, 20, 36, 72, 32, 50, 20, 44, 70, 12, 26])

#Generate a boxplot of sales
    #Note: boxplot() method, will automatically sort the given range(values) in ascending order,
                           # also, it will find the 25%, median, & 75% of the given range.
plt.boxplot(sales)

#Giving, title, xlabel, and ylabel to boxplot
plt.title("Title: Sales Box Plot")
plt.xlabel("Number")
plt.ylabel("Sales Values")

plt.show()
```

```python
# 2: Create a box plot having figure size width(x-axis)=10, length(Y-axis)=7

import numpy as np
import matplotlib.pyplot as plt

sales = np.array([6, 7, 13, 20, 17, 39, 41, 25, 43, 51, 49, 62])

#Generate a boxplot of sales with figsize = (6,5)
plt.figure(figsize=(10,7))
plt.boxplot(sales)
plt.show()
```

```python
# 3: Create a box plot using "students.csv" CSV file

import pandas as pd
import matplotlib.pyplot as plt

csv_file = pd.read_csv("students.csv")

plt.title("Overall Class Score")
plt.ylabel("Students Score")
plt.boxplot(csv_file['total_score'])
plt.show()
```

```python
# 4: Create a box plot using CSV and Dataframe
# which display using dot('o'), whether or not a mean values of various categories (boxes) are significantly different or not

import pandas as pd
import matplotlib.pyplot as plt

#Reading a csv file and storing that data into stuData
stuData = pd.read_csv("students.csv")

#Creating dataframe and storing CSV file data into it.
df = pd.DataFrame(stuData)

#Create a boxplot using DataFrame
df.plot(kind='box')


plt.title("Students Subject wise marks")
plt.xlabel("Subjects")
plt.ylabel("Score")
plt.show()
```

```python
# 5. We have a csv file (banks.csv) in which we have columns (acno, cname, account_type, balance). Draw a boxplot

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("banks.csv")

#Create a boxplot using csv object, here by(x-axis), column(y-axis)
df.boxplot(by='account_type', column=['balance'], color='blue')

plt.title("Bank Record")
plt.xlabel("Account Types")
plt.ylabel("Balance")
plt.show()
```

```python
# 6. Draw boxplot using loan.csv file and IPython library

import pandas as pd
import matplotlib.pyplot as plt
from IPython import get_ipython

get_ipython().run_line_magic('matplotlib', 'inline')

df = pd.read_csv("loan.csv")  #Housing loan and Vehical loan

df.boxplot(by='Loan_Type', column=['Amount'], color='red')
```
