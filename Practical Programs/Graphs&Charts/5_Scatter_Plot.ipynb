```python
#Scatter Plot
# Program 1: Basic scatter plot showing the Houses and their prices.

import matplotlib.pyplot as plt

#prices (y-axis) of 1bhk, 2bhk, 3bhk, 4bhk, 5bhk, and 6bhk houses (x-axis)
houses = ["1BHK", "2BHK", "3BHK", "4BHK", "5BHK", "6BHK"]
prices = [55,65,80,90,140,210]

#contruct a scatter plot x-axis denotes "Houses", and y-axis denotes "Prices"
plt.scatter(houses, prices)

plt.show()
```

```python
#Scatter Plot
#Program 2: Female Height v/s Weight scatter plot.

import matplotlib.pyplot as plt
import pandas as pd

# Read the data from file
df = pd.read_csv("height_weight.csv")

# Figure Size
fig = plt.subplots(figsize=(10,6))

# Construct figure with labels
fig.scatter(x = df['F_height'], y=df['F_weight'])
plt.xlabel("Height of Female")
plt.ylabel("Weight of Female")

#Display figure
plt.show()
```

```python
#Scatter Plot
#Program 3: Sales v/s Discount Scatter plot

import numpy as np
import matplotlib.pyplot as plt

discount = np.array([30,20,50,40,10])
salesInRs = np.array([40000, 45000, 68000, 50000, 60000])
size = discount*10

# Contuct a scatter plot where,
    # x-axis => Dicsount
    # y-axis => Sales in RS
    # size = discount*10
    # design => start (*)
    # color and edgecolor
plt.scatter(x=discount, y=salesInRs, s=size, marker='*', color='red', edgecolor='yellow')

plt.title("Sales v/s Dicsount")
plt.xlabel("Discount Offered")
plt.ylabel("Sales In Rs")

plt.show()
```

```python
#Scatter Plot
# Program 4: Scatter plot using seaborn python library

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file = pd.read_csv('diamonds.csv')

fix = plt.subplots(figsize=(10,6))

sns.scatterplot(x=file['carat'], y=file['depth'], hue=file['cut'])

plt.title("Carrot Depth")
plt.xlabel("carat")
plt.ylabel("Depth")

plt.show()
```
