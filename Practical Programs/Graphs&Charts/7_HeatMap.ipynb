```python
# HeatMap Plots

#Note: For small data we use Pie chart and Bar chart and For large set of data we used Scatter plot and heat chart

#Program 1: Create a sample HeatMap
#Program 2: Create a sample HeatMap with annotaion for each box
#Program 3: Create a sample HeatMap with specific hue
```

```python
#Program 1: Create a sample HeatMap

import seaborn as sns
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.random((6,5)), columns=["a", "b", "c", "d", "e"])

sns.heatmap(df)

sns.heatmap(df, annot=True) # annot_kws={"size":10}
```

```python
#Program 2: Create a sample HeatMap with annotaion for each box

import seaborn as sns
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.random((6,5)), columns=["a", "b", "c", "d", "e"])

#Note: when using annot=Ture, the color combination is used automatically

sns.heatmap(df, annot=True) # annot_kws={"size":10}
```

```python
#Program 3: Create a sample HeatMap with specific hue

import seaborn as sns
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.random((10,10)), columns=["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"])

#Note: when using annot=Ture, the color combination is used automatically

#Plot using a color palette
sns.heatmap(df)
```
