```python
# 1. Creating a pie chart (Basic)

#matplotlib.pyplot is a library need for drawing various graph
import matplotlib.pyplot as plt
#numpy is a library use fo arrays purpose, without numpy we can't create array in python
import numpy as np

#Creating an array, which is storing the value for pie chart
y = np.array([10, 20, 30, 40])

#Contructing the pie chart
plt.pie(y) #OR, plt.pie((10,20,30,40))
#Display the pie chart
plt.show() #show() display all open figures
```

```python
# 2. Creating a pie chart with labels

import matplotlib.pyplot as plt
import numpy as np

points = np.array([10, 20, 30, 40])
label = ["Label1", "Label2", "Label3", "Label4"]

#Contructing the pie chart, having values(points array) and labels(label list)
plt.pie(points, labels=label)
#OR, plt.pie(points, labels=("Label1", "Label2", "Label3", "Label4"))

plt.show()
```

```python
# 3. Creating a pie chart with legends(index) and title

import matplotlib.pyplot as plt
import numpy as np

points = np.array([10, 20, 30, 15])
label = ["L1", "L2", "L3", "L4"]

#Contruct the pie chart
plt.pie(points, labels=label)
#specify the legend and title for pie chart
plt.legend(title="Title Here")

plt.show()
```

```python
# 4. Creating a pie chart with legends(index) positing at lower right.

import numpy as np
import matplotlib.pyplot as plt

range = np.array([10, 20, 30, 40])
name = ["N1", "N2", "N3", "N4"]

#contruct a pie chart having value(range) and labels(name)
plt.pie(range, labels=name)

#create a legend for pie chart, display location = "Lower right" (By default it is upper-right)
plt.legend(title="Names", loc="lower right")

plt.show()
```

```python
# 5. Creating a pie chart with explode wedge

import matplotlib.pyplot as plt

# Array is not always requried, but it is recommended, when we want to have a collection of one type of value only.
#Here, we are using list only.
percentage = [50, 20]

#Contructing a pie chart, with 
    # value i.e. percentage,
    # labels / Index for the pie chart
    # explode (explode define the (position, value) that you want to show as unique/separate)
plt.pie(percentage, labels=("Like", "Hate"), explode=(0,0.1))

plt.legend(title="Math Status")
plt.show()
```

```python
# 6. Creating a pie chart, where section is design by user define colors

import matplotlib.pyplot as plt
import numpy as np

range = np.array([10, 15, 60, 35])
label = ["C1", "C2", "C3", "C4"]
color_name = ["pink", "grey", "white", "brown"]

print("Pie chart with \"User Define colors\" ")
#contructing a pie chart with value, lables, and colors
plt.title("Pie Chart with User \"Define Colors\" ")
plt.pie(range, labels=label, colors=color_name)
#display the pie chart
plt.show()
```

```python
# 7. Create a pie chart, having startangle=90.

import numpy as np
import matplotlib.pyplot as plt

value = np.array([50,20,30,15])
label = ["Ten", "Twenty", "Thirty", "Fifteen"]
color = ["white", "red", "grey", "yellow"]

print("Pie chart starting at 90 Degree, Anti-Clock wise")
#contruct a pie chart which start from 90 degree, in anti-clock wise
plt.pie(value, labels=label, colors=color, startangle=90)

plt.show()
```

```python
# 8. Create a pie chart, having two part separated/cut i.e explode

import numpy as np
import matplotlib.pyplot as plt

range = np.array([200,145, 100, 175])
label = ["Area", "Width", "Length", "Volumne"]

#Here we define a list which define the separation width,
# here, 0.2 -> Area, 0 -> width, 0.3 -> Length, 0 -> volumne
separate_part = [0.2, 0, 0.3, 0]

#contruct a pie chart
plt.pie(range, labels=label, explode=separate_part)
plt.show()
```

```python
#Practical Programs
# 9. Draw pie chart using numpy and csv file

import matplotlib.pyplot as plt
import numpy as np

label = ["Informatvie", "Entertainment", "Sports", "News"]
point = np.array([10, 50, 25, 15])

print("Pie chart using NumPy Arrays:")
plt.title("PieCharts")
plt.pie(point, labels=label, autopct="%1.2f%%")
plt.show()
```

```python
#Practical Programs
# 10. Draw a pie chart having labels, legend, title to the legend, and user-define colors
# Given, Expenses = Rent, Grocery, Transport, Current, School Fee, Savings
# and,   Amount   = 7000, 3000, 800, 300, 2000, 1900

import numpy as np
import matplotlib.pyplot as plt

expenses = np.array(["Rent", "Grocery", "Transport", "Current", "School Fee", "Savings"])
amount = np.array([7000, 3000, 800, 300, 2000, 1900])
color = ["white", "grey", "pink", "red", "orange", "green"]

plt.pie(amount, labels=expenses, colors=color)
plt.legend(title="Expenses Amt")
plt.show()
```

```python
#Practical programs
# 11. Draw a pie chart displaying the numbers of gold-medals win by specific country using CSV file

#import pandas library for CSV file operation
import pandas as pd

#import matplotlib.pyplot library for drawing charts
import matplotlib.pyplot as plt

#read medals.csv file
file_csv = pd.read_csv("medals.csv")

countryName = file_csv['country']
totalGoldMedal = file_csv['gold_medals']

plt.pie(totalGoldMedal, labels=countryName, startangle=90, autopct='%1.0f%%')
plt.show()
```
