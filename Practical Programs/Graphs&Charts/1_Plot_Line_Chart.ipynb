```python
# 1. Drawing a line/plot chart

#import matplotlib.pyplot for visual effect
import matplotlib.pyplot as plt

#import numpy for arrays
import numpy as np

# Create two arrays
xpoints = np.array([0, 4])
ypoints = np.array([0, 250])

# Ploting a grpah with points as (0,0) and (4, 250)
plt.plot(xpoints, ypoints)
plt.show()
```

```python
#2. Drawing a plot char with dot instead of line

#import matplotlib.pytplot for visual effect
import matplotlib.pyplot as plt

#import numpy for arrays
import numpy as np

#Create two arrays
xpt = np.array([2,10])
ypt=np.array([5,50])

#Plot the graph, by default its show line, but here we explicitly mention "o"
plt.plot(xpt, ypt, "o")
plt.show()
```
