```python
# Using CSV library
# Create a CSV file and puts records in it.

#import csv library
import csv
import pandas as pd

# open file using with statement
with open("doc.csv", 'w', newline='') as file:
    # csv.writer class is used to insert data to the CSV file. 
    # This class returns a writer object which is responsible for converting the user’s data into a delimited string.
    csvwriter = csv.writer(file)

    # csv.writer class provides two methods for writing to CSV. A) writerow() & B) writerows().
    csvwriter.writerow(["Dno", "Dname", "DLoc"])
    csvwriter.writerow([101, "Doc1", "Loc1"])
    csvwriter.writerow([102, "Doc2", "Loc2"])

fread = pd.read_csv("doc.csv")
fread
```

```python
# Using CSV library, Open append data to existing CSV file

#import csv
import csv

#Open CSV file
with open("doc.csv", 'a', newline="") as file:
    csvWriter = csv.writer(file)
    csvWriter.writerow([103, "Doc3", "Loc3"])
    csvWriter.writerow([104, "Doc4", "Loc4"])

import pandas as pd
fo = pd.read_csv("doc.csv")
fo
```

```python
# Using CSV Library, Create/open a file in write mode and write multiple rows at a time into the file

import csv

#Create list
carsList = [ ["Cno", "Cname", "Ccolor"], [101, "Car1", "Color1"], [102, "Car2", "Color2"] ]

#open file in write mode
with open("cars.csv", "w", newline="") as file:
    csvWriter = csv.writer(file)
    csvWriter.writerows(carsList)

import pandas as pd
fo = pd.read_csv("cars.csv")
fo
```

```python
# Practical
# Fetch data from one csv file and add it into new csv file

import csv
import pandas as pd

ListF1=[]
f1 = open("emp.csv")
for x in f1:
    ListF1.append(x)

ListF1

with open("emp3.csv", 'w+', newline="") as file2:
    csvWriter = csv.writer(file2)
    csvWriter.writerows(ListF1)

f2 = pd.read_csv("emp2.csv")
f2
```
